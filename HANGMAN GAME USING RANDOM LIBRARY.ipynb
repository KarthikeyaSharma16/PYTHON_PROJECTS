```python
import random as r

#PICTURE
hangman_boii = [''' 
        +---+
            |
            |
            |
           ===''', ''' 
        +---+
        o   |
            |
            |
           === ''',''' 
        +---+
        o   |
        |   |
            |
           ===''', ''' 
        +---+
        o   |
        |\  |
            |
           === ''',''' 
        +---+
        o   |
       /|\  |
            |
           ===''', ''' 
        +---+
        o   |
       /|\  |
         \  |
           === ''',''' 
        +---+
        o   |
       /|\  |
       / \  |
           === ''']

#HINT
print("Genre: Processor and FPGA manufacturing companies")
print("HANGMAN",hangman_boii[0])

#LIST OF WORDS
list_of_words = ['Intel','Altera','ARM','Xilinx','AMD','Lattice','Microchip','Atmel','Arduino','Samsung','Apple','Qualcomm','Mediatek']
Word_to_be_guessed = list_of_words[r.randint(0,len(list_of_words)-1)]

#BLANKS
blanks = []
for ii in range(len(Word_to_be_guessed)):
    blanks.append('_')
print(' '.join(blanks))

already_guessed = []
correct_guess = []
missing_letters = []


for ii in Word_to_be_guessed:
    missing_letters.append(ii)

def Input():
    players_input = input("Enter a letter: ")
    return players_input

def CompUserInput(players_input):
    count = 0 
    #if there is more than one occurance of the players input, create a for loop, append the total number of repeated letters in the list correct guess and remove the same number of letters from the list missing_letters
    if players_input in Word_to_be_guessed:
        for ii in range(len(Word_to_be_guessed)):
            if Word_to_be_guessed[ii] == players_input and players_input in missing_letters:
                correct_guess.append(players_input)
                missing_letters.remove(players_input)
        for ii in range(len(Word_to_be_guessed)):
            if players_input == Word_to_be_guessed[ii]:
                blanks[ii] = players_input
                r = ' '.join(blanks)
                
        for ii in blanks:
            if ii.isalpha():
                count += 1
            if count == len(Word_to_be_guessed):
                break
        print(r)
        
    elif players_input not in Word_to_be_guessed:
        already_guessed.append(players_input)
        if len(already_guessed) == 1:
            print(hangman_boii[1])
        else:
            print(hangman_boii[len(already_guessed)+1])

help_player = 0
while True:
    players_input = Input()
    CompUserInput(players_input)
    if len(already_guessed) == 5:
        print("Gameover!")
        break
    if len(correct_guess) == len(Word_to_be_guessed):
        print("You won!")
        break
    if len(already_guessed) >= 2 and help_player == 0:
        assistance = input("Do you want assistance? Type y (Yes) or n (No)")
        if assistance == 'y':
            help_player += 1
            computer_input = missing_letters[r.randint(0,len(missing_letters)-1)]
            CompUserInput(computer_input)
        elif assistance == 'n':
            continue
```
