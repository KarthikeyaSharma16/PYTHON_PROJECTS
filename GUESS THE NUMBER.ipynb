```python
import random as r
x = r.randint(1,100)
def randi():
    guess = int(input("Enter a number to guess: "))
    return guess

no_of_attempts = 0
guess = randi()

while True:
    if guess == x:
        break
    elif guess != x:
        if guess > x:
            print("Go lower!")
        elif guess < x:
            print("Go higher!")
        no_of_attempts += 1
        guess = randi()
    
print(f"You have guessed the number! Number of attempts taken to guess the number is {no_of_attempts}")
```
