```python
import numpy as np
import matplotlib.pyplot as plt
import random
plt.style.use("seaborn-poster")
n = int(input("Enter the number of steps: "))
x = np.zeros(n)
y = np.zeros(n)
direction = ["NORTH","SOUTH","EAST","WEST"]
for i in range(1,n):
    step = random.choice(direction)
    if step == "EAST":
        x[i] = x[i-1] + 1
        y[i] = y[i-1]
    elif step == "WEST":
        x[i] = x[i-1] - 1
        y[i] = y[i-1]
    elif step == "NORTH":
        x[i] = x[i-1] 
        y[i] = y[i-1] + 1
    else:
        x[i] = x[i-1] 
        y[i] = y[i-1] - 1
plt.title("2-D random Walk")
plt.plot(x,y,color = "#ffabcd", label = "Random Walk in 2-D")
plt.legend(loc = "upper right", shadow = True)
plt.grid(True)
plt.tight_layout()
plt.show()
```
