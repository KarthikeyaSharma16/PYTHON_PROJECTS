```python
students_list = list()
n = int(input("Enter the number of students: "))
for student in range(n):
    student_name = input("Enter the name of the student: ")
    students_list.append(student_name)
print(f"The list of students are {students_list}")
for student_name in students_list:
    print(f"{student_name}:")
    student = {"Theory" : [float(input("Enter CT1 marks (out of 25): ")),float(input("Enter CT2 marks (out of 50): ")),float(input("Enter CLAT3 marks (out of 100): "))], "Lab":[float(input("Enter Lab 1 mark (out of 30): ")),float(input("Enter Lab 2 mark (out of 30): ")),float(input("Enter Lab 3 mark (out of 30): ")),float(input("Enter Lab 4 mark (out of 30): ")),float(input("Enter Lab 5 mark (out of 30): ")),float(input("Enter Lab 6 mark (out of 30): ")),float(input("Enter Lab 7 mark (out of 30): ")),float(input("Enter Lab 8 mark (out of 30): ")),float(input("Enter Lab 9 mark (out of 30): ")),float(input("Enter Lab 10 mark (out of 30): "))], "Assignments":[float(input("Enter Assignment 1 marks: ")),float(input("Enter Assignment 2 marks: ")),float(input("Enter Assignment 3 marks: ")),float(input("Enter Assignment 4 marks: "))]}
    print(student)
    for (k, v) in student.items():
        total = 0
        for marks in v:
            total = total + marks
            if k == "Theory":
                theory_final = (4/7)*total*0.7
            elif k == "Lab":
                lab_final = (1/3)*total*0.2
            elif k == "Assignments":
                assignment_final = 2.5*total*0.1
    total_mark = theory_final + lab_final + assignment_final
    print(total_mark)
    if total_mark >= 90:
        print(f"{student_name} scored A grade")
    elif total_mark >= 80:
        print(f"{student_name} scored B grade")
    elif total_mark >= 70:
        print(f"{student_name} scored C grade")
    elif total_mark >= 60:
        print(f"{student_name} scored D grade")
    elif total_mark <= 60:
        print(f"{student_name} scored E grade")
```
